```python
import math
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import norm

plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
##The routine is to create standard distribution to calculate probabilities
def draw_normal_distribution(col_val, range_val=0.90, color_val="lime"):
    total_mean = data["Total"].mean()
    total_sigma = data["Total"].std()
    mean = data[col_val].mean() / total_mean
    sigma = data[col_val].std() / total_sigma

    x = np.arange(-4, 4, 0.001)
    y = norm.pdf(x,0,1)

    range_val = range_val / 2
    lower_bound = mean - (mean * range_val)
    upper_bound = mean + (mean * range_val)
    z1 = (lower_bound - mean) / sigma
    z2 = (upper_bound - mean) / sigma

    fig, ax = plt.subplots(figsize=(8,6))
    ax.set_xticklabels([str(round(((x * sigma) + mean)*1000,2))+" [" + str(x) + "] " for x in range(-5, 5)], rotation=90)
    ax.plot(x,y)

    fill_x = np.arange(z1, z2, 0.001)
    fill_y = norm.pdf(fill_x, 0, 1)
    ax.fill_between(fill_x, fill_y, 0, alpha=0.4, color="lime")

    ax.text(-3.1,0.03, "sigma = {}".format(sigma), rotation=90)
    ax.text(-2.8,0.03, "mu = {}".format(mean), rotation=90)
    ax.set_xlabel("Distribution of {}".format(col_val))
    p = str(round((norm.cdf(z2, 0, 1) - norm.cdf(z1, 0, 1)) *100,2))
    ax.text(z1+0.1,0.10,"p="+p +"%", rotation=90)
    ax.set_title("What is the probability that {}% of the pea population has a {}?".format(range_val*200,col_val))
    plt.show();

def draw_total_normal_distribution(col_val, color_val="lime"):
    total_mean = data["Total"].mean()
    total_sigma = data["Total"].std()
    mean = total_mean
    sigma = total_sigma

    x = np.arange(-4, 4, 0.001)
    y = norm.pdf(x,0,1)

#     range_val = range_val / 2
    lower_bound = -1
    upper_bound = 1
    z1 = lower_bound
    z2 = upper_bound

    fig, ax = plt.subplots(figsize=(8,6))
    ax.set_xticklabels([str(round(((x * sigma) + mean),2))+" [" + str(x) + "] " for x in range(-5, 5)], rotation=90)
    ax.plot(x,y)

    fill_x = np.arange(z1, z2, 0.001)
    fill_y = norm.pdf(fill_x, 0, 1)
    ax.fill_between(fill_x, fill_y, 0, alpha=0.4, color="lime")

    ax.text(-3.1,0.03, "sigma = {}".format(sigma), rotation=90)
    ax.text(-2.8,0.03, "mu = {}".format(mean), rotation=90)
    ax.set_xlabel("Distribution of {}".format("Total Pea population over 30 generations with {}".format(col_val)))
    p = str(round((norm.cdf(z2, 0, 1) - norm.cdf(z1, 0, 1)) *100,2))
    z1_val = round(((z1 * sigma) + mean),2)
    z2_val = round(((z2 * sigma) + mean),2)
    ax.text(z1+(z2/2),0.10,"p="+p +"%", size=12)
    ax.set_title("The probability that the total number of peas in each generation fall between {} and {} (1 sigma) is {}%".format(z1_val, z2_val, p))
    plt.show();
```
